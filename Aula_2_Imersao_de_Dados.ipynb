```python
#para criar novas colunas, mesmo método
#quebrando a coluna valor em Moeda, Valor anuncio e Tipo anuncio
dados[["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados["Valor"].str.split(expand = True)
dados.head()
```

```python
#eliminar o tipo de anuncio que não é vazio
#ver se a coluna tipo anuncio é vazio
dados["Tipo_anuncio"].isnull()
```

```python
#só tipo q isnull, é nulo
dados[dados["Tipo_anuncio"].isnull()]["Tipo_anuncio"].unique()
```

```python
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
#Valor_anuncio está tipo objeto
#precisa converter o ponto para poder tirar a média
#mesmo sendo string ele vai tentar tirar a média - mas, ele trava no ponto "."
dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio_float"].plot.hist(bins = 50)
```

```python
pd.set_option("display.precision", 2)
#formato tipo float com duas casas decimais à direita
pd.set_option('display.float_format', lambda x: '%.2f' % x)

dados_vendas.describe()
```
