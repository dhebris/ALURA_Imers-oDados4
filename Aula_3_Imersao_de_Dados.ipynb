```python
#Valor Metro quadrado por bairro2
#coluna com o valor criada "Valo-m2"
dados_vendas["Valor_m2"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
dados_vendas.head()
```

```python
#Agrupar por bairrom chama a função média
dados_vendas.groupby("Bairro").mean()
```

```python
#soma
dados_bairro = dados_vendas.groupby("Bairro").sum()
dados_bairro
```

```python
#criar coluna valor metro quadrado bairro
dados_bairro["Valor_m2_bairro"] = dados_bairro["Valor_anuncio_float"]/dados_bairro["Metragem"]
dados_bairro
```

```python
#.loc busca pela string -- SE For buscar por Vila Mariana apenas vai dar erro, pq o index agora é o bairro
#qdo usa o groupby - tudo q vc agrupa vira o index dps, por isso precisa usa o método acima - precisa resetar para usar da forma de index
dados_bairro.loc["Vila Mariana"]
```

```python
#retornar os bairros que mais têm imóveis
dados_vendas["Bairro"].value_counts()
```

```python
#slice
dados_vendas["Bairro"].value_counts().index
```

```python
#dez primeiros
top_bairros = dados_vendas["Bairro"].value_counts()[:10].index
```

```python
#plotar em um gráfico - https://seaborn.pydata.org/generated/seaborn.barplot.html
#para utilizar,antes precisa resetar o index dos bairros - precisa virar uma coluna
#repare no index antes da coluna bairro -  o index virou uma coluna
```

```python

dados_bairro.reset_index()
dados_bairro
```

```python
#precisa associar usando o TRUE
dados_bairro.reset_index(inplace = True)
dados_bairro
```

```python
dados_bairro.query("Bairro in @top_bairros")
```

```python
plt.figure(figsize=(10, 8))
ax = sns.barplot(x="Bairro", y="Valor_m2_bairro", data=dados_bairro.query("Bairro in @top_bairros"))
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros"), x="Bairro", y="Valor_anuncio_float")
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros & Metragem < 30000"), x="Bairro", y="Metragem")
ax.tick_params(axis='x', rotation=45)
plt.show()
```
